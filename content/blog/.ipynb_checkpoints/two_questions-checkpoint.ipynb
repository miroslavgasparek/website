- summary: Two important questions that can guide your thinking process.

In the summer of 2018, I spent two months in the research group of Professor Drew Endy at Stanford University. This time was priceless. Yes, I acquired many new skills, learned a lot about the synthetic biology, and made a lot of friends.

However, some of the most important aspects of my internship were the opportunities to learn and talk about general concepts regarding the science, research, and their meaning. And one of the main things I learned from Drew was to ask the questions: **Why?** and **Why does it matter?** when trying to find out if the things I am planing to do are meaningful.

During our first discussion, on the very first day of my internship, I presented my - what I considered to be - well-polished and detailed research proposal. It had the description of three different potential synthetic biology research projects, which I considered to be well-thought-out, detailed, and somehow manageable (Yes, I was even more naive then!). 

However, when I described my ideas about i) external control of the cellular processes, ii) computational modelling of the "communication" between synthetic cells, and iii) interconnection of the cellular subsystems, Drew asked me two simple questions:

* Why do you want to do these projects?
* Why do they matter?

And I felt that I did not have the satisfactory answers to these questions.

I felt that the brutally honest answers would be along “because I have done similar things, and this is the extension of what I have done” or “I have read the papers in which people did almost exactly the same things…”

But I somehow felt that these motivations are not completely “right”. And to my silence, Drew said: "What you proposed are projects that you can try to do. However, the questions that you should ask are: Why do the things that you want to do matter? Why, at the end of the summer, when you are done with your internship, should anyone care about what you did? What is the meaning of the problem that you are planning to solve?”

And then he suggested that I should try to identify the functions of the unknown essential genes in the smallest living organism - a  topic that deserves its own blog (or perhaps textbook, and then library). He then explained why it is important to find out what the functions of essential genes are. And I understood that this is such a fundamental scientific problem, with so many potential applications, that *everyone* would benefit if we solved this problem.

This was a surprising twist in my plans - I thought that I would work on a computational project, with some experimental work - instead, I spend pretty much the whole summer in the lab, working towards the realization of the assay that could help us elucidate the function of these mysterious genes (and screwing things up!). It was quite challenging - but I learned incredible amount of interesting things!

And that changed the way I would like to look at the projects I choose. I am trying to ask: Why do I want to do this? And why does it matter? Will this project, if successful, be something important for the wider community, will it help me to gain some unique experience, or perhaps will it truly solidify my skills? And will people actually care?

I suggest that you also try to ask these questions when deciding what to do, or which project to choose. It does not have to be a research project - these questions work well for start-up ideas, as well as for the "Netflix vs. hiking trip" dilemmas.

So why do you want to do this? And why does it matter?
